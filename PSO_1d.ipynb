a)The trajectory graph looks clear because the lines of particles don't intersect with each other very often, and the lines looks more gentle. Relatively small vmax, which is good for exploitation, but it is easily to fall in the trap of local optima.

b)The trajectory graph looks like a chaos because lines of particles intersect with each other very often, due to the rapid changing slope. Relatively large vmax, which is good for exploration, but it is easily to overshoot from ("fly over") the true global optimum.

c)Rule of thumb of $v_{max}:(x_{max}-x_{min})/n_{steps}*20$
